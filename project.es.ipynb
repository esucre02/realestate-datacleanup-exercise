#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
def normalize(group):
    return (group - group.min()) / (group.max() - group.min())

df_cinturon_sur['price_norm'] = df_cinturon_sur.groupby('level5')['price'].transform(normalize)

# 3. Preparar los datos para el histograma múltiple
data_to_plot = [
    df_cinturon_sur[df_cinturon_sur['level5'] == p]['price_norm'] 
    for p in poblaciones_del_sur
]

# 4. Trazar los histogramas
plt.figure(figsize=(12, 7))
plt.hist(data_to_plot, bins=15, label=poblaciones_del_sur, alpha=0.7, density=True)

plt.title('Histogramas de Precios Normalizados (Cinturón Sur)', fontsize=15)
plt.xlabel('Precio Normalizado (0 = Mínimo, 1 = Máximo)', fontsize=12)
plt.ylabel('Densidad de frecuencia', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.show()

```

Al normalizar los precios podemos ver que la mayor cantidad de casas en el cinturon del sur son mas baratas, sin importar el pueblo. Podemos ver que Getafe sigue siendo el que en promedio tiene las casas mas caras con respecto al resto.

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
fig, axes = plt.subplots(2,2, figsize=(15, 10))
fig.suptitle('Relación Superficie vs Precio en el Cinturón Sur')

axes_flat = axes.flatten()

for i, poblacion in enumerate(poblaciones_del_sur):

    data = df_cinturon_sur[df_cinturon_sur['level5'] == poblacion]
    
   
    axes_flat[i].scatter(data['surface'], data['price'], alpha=0.5, s=20, color=f'C{i}')
    
    
    axes_flat[i].set_title(poblacion, fontsize=14)
    axes_flat[i].grid(True, linestyle='--', alpha=0.6)
    
    
    if i >= 2: axes_flat[i].set_xlabel('Superficie (m²)')
    if i % 2 == 0: axes_flat[i].set_ylabel('Precio (€)')

plt.tight_layout()
plt.show()

```

```python
import plotly.express as px


# Limpiamos comas y convertimos a numérico. 
df_cinturon_sur['latitude'] = pd.to_numeric(df_cinturon_sur['latitude'].astype(str).str.replace(',', '.'), errors='coerce')
df_cinturon_sur['longitude'] = pd.to_numeric(df_cinturon_sur['longitude'].astype(str).str.replace(',', '.'), errors='coerce')

# Eliminamos filas que se hayan quedado sin coordenadas tras la limpieza
df_mapa = df_cinturon_sur.dropna(subset=['latitude', 'longitude'])

#  TRANSFORMACIÓN A DICCIONARIO 
datos_dict = df_mapa[['latitude', 'longitude', 'level5', 'price']].to_dict('records')

#  CREACIÓN DEL MAPA
fig = px.scatter_map(
    datos_dict, 
    lat="latitude", 
    lon="longitude", 
    color="level5",
    hover_name="level5",
    hover_data={"price": True, "latitude": False, "longitude": False},
    zoom=10, 
    center={"lat": 40.32, "lon": -3.75}, # Centro manual en el sur de Madrid
    height=600,
    title="Mapa del Cinturón Sur de Madrid"
)

fig.update_layout(map_style="open-street-map")
fig.show()
```
